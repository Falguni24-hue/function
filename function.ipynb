Q6 **Explain concept of generator in python and how they are defined.**

Generator is a function that yields value one by one using the yield keyword instead of returning all value using **return** Each time the generator’s __next__() method is called.Generators are a type of iterator, so they implement the __iter__() and __next__() methods, but unlike regular iterators, they are created using a function with yield.


```python
def square_number(n):
  for i in range(n):
    yield i**2
```

```python
square=square_number(5)
```

```python
list(square_number(5))
```

```python
next(square)
```

```python
next(square)
```

Q7 **what are the advantage of generator over regular fuction.**

Generators produce values one at a time, only when needed. They don’t generate and store all values in memory at once, making them much more memory-efficient, especially when working with large datasets.generators only yield one item at a time, they consume very little memory even for large or infinite sequences.**Regular functions** return all values at once.which requires storing all the values in memory. This can quickly become problematic when the dataset is large.

```python
def square_number(n):   #code using regular function
  result = []
  for i in n:
    result.append(i**2)
  return result
```

```python
square_number([5,20,45,23])
```

```python
def square_number(n):    #code using generator function
  for i in range(n):
    yield i**2
```

```python
list(square_number(5))
```

```python
square=square_number(5)
```

```python
next(square)
```

Q9 **Explain the purpose and use of map function in python**

The map() function in Python is a built-in function that allows you to apply a given function to each item in an iterable (like a list, tuple, etc.) and return an iterator that produces the results. Essentially, it "maps" a function to every element of an iterable and yields the result of applying that function to each element.



```python
# map function with regular function
words = ["pw","skills"]
list(map(str.capitalize,words))
```

```python
# map function with lambda function.
words = ["pw","skills"]
list(map(lambda str : str.capitalize(),words))
```

Q10 **what is difference between 'map()','reduce()'and 'filter()' functions in python?**

map() applies a given function to each item in an iterable (such as a list or tuple) and returns an iterator that yields the results.

**reduce()** applies a binary function cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single result.**reduce() import form functools library.**

**filter()** applies a function to each item of an iterable and returns a filtered iterator containing only the items for which the function returns True.

```python
#map()
l = [1,2,3,4,5]
list(map(lambda x : x**2,l))
```

```python
#reduce()
from functools import reduce
l = [1,2,3,4,5]
reduce(lambda x ,y : x+ y**2 ,l)
```

```python
#filter()
l =[1,2,2,2,3,4,5,6]
list(filter(lambda x : x%2==0,l))
```

Q1 **Write a python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def sum_even_numbers(numbers):
  even_numbers =(num for num in numbers if num % 2 ==0)
  return sum(even_numbers)
```

Q5 **Create a iterator class in python that generate the fibonacci sequence up to a specified number of terms.**

```python
def fib(n):
    a, b = 0, 1
    result = []
    for i in range(n):
        result.append(a)
        a, b = b, a + b  # Update a and b for the next Fibonacci number
    return result
```

```python
fib(12)
```
